```python
import os
import logging
import tifffile
from sam_module import SAMProcessor, SAMVisualizer
from tqdm.notebook import tqdm

# Set up logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
```

## Initialize SAM Processor and Visualizer

```python
# Configuration
sam_checkpoint = "sam_vit_h_4b8939.pth"  # Path to your SAM checkpoint
output_folder = "processed_frames"
os.makedirs(output_folder, exist_ok=True)

# Initialize processor and visualizer
processor = SAMProcessor(sam_checkpoint)
visualizer = SAMVisualizer()
```

## Process TIFF File

```python
def process_tiff_frames(tiff_path: str, num_frames: int = 10):
    """Process specified number of frames from a TIFF file."""
    with tifffile.TiffFile(tiff_path) as tif:
        total_frames = len(tif.pages)
        num_frames = min(num_frames, total_frames)
        print(f"Processing {num_frames} frames out of {total_frames} total frames")
        
        processed_images = []
        processed_masks = []
        
        for frame_idx in tqdm(range(num_frames)):
            # Load and process frame
            frame = processor.load_tiff_frame(tiff_path, frame_idx)
            original, masks = processor.process_frame(frame)
            
            # Store results
            processed_images.append(original)
            processed_masks.append(masks)
            
            # Save individual frame results
            fig = visualizer.visualize_masks(original, masks)
            if fig:
                output_path = os.path.join(output_folder, f"frame_{frame_idx:03d}_output.png")
                visualizer.save_visualization(fig, output_path)
        
        return processed_images, processed_masks
```

## Run Processing

```python
# Specify your TIFF file path and number of frames to process
tiff_path = "AR1_K_mA.tif"  # Replace with your TIFF file path
num_frames = 10

# Process the frames
processed_images, processed_masks = process_tiff_frames(tiff_path, num_frames)

# Display results in a grid
fig = visualizer.display_multiple_frames(processed_images, processed_masks)
visualizer.save_visualization(fig, os.path.join(output_folder, "all_frames_summary.png"))
```
